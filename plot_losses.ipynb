```python
fig,(ax1,ax2) = plt.subplots(ncols=2)

fig.set_figwidth(12)
fig.set_figheight(5)

ax = ax1
x0 = [8, 16, 32, 64, 128, 256]
x = np.log(x0)
y1 = [97.79, 97.2, 96.21, 94.88, 92.8, 89.4] # 3 epochs
y2 = [98.03, 98.06, 97.51, 96.53, 94.93, 91.70] # 6 epochs
y3 = [98.43, 98.26, 97.9, 97.2, 96.21, 94.25] # 10 epochs

_=ax.plot(x,y1,'-^',label='3 epochs')
_=ax.plot(x,y2,'-o',label='6 epochs')
_=ax.plot(x,y3,'-s',label='10 epochs')
_=ax.set_title("(a) Accuracy vs mini-batch size",size=18)
_=ax.set_xlabel("Mini-batch size",size=16)
#_=ax.set_yticks(y)
_=ax.set_yticklabels(ax.get_yticks(), size=14)
_=ax.set_xticks(x)
_=ax.set_xticklabels(x0, size=16)

_=ax.legend(prop={'size':16})

ax = ax2

x = [1,2,3,4,5,6] #epochs
y1 = [97.06, 97.48, 97.20, 97.61, 96.69, 97.73] #m0.5, e0.05
y2 = [96.60, 96.92, 97.25, 97.35, 97.65, 97.86] #m0.9, e0.01
y3 = [96.75, 97.48, 97.04, 97.60, 96.32, 97.42] #m0.7, e0.03
y4 = [94.55, 96.56, 97.15, 97.48, 97.79, 98.06] #m0,   e0.01
y5 = [96.42, 97.46, 97.86, 97.90, 98.03, 98.25] #m0.5, e0.01
y6 = [95.64, 96.98, 97.57, 97.86, 97.94, 98.28] #m0.3, e0.01

_=ax.plot(x, y1, '-^', label="lr=0.05, m=0.5")
_=ax.plot(x, y2, '-o', label="lr=0.01, m=0.9")
_=ax.plot(x, y3, '-+', label="lr=0.03, m=0.7")
_=ax.plot(x, y4, '-v', label="lr=0.01, m=0")
_=ax.plot(x, y5, '-x', label="lr=0.01, m=0.5")
_=ax.plot(x, y6, '-s', label="lr=0.01, m=0.3")

_=ax.set_title("(b) Accuracy vs lr and momentum", size=18)
_=ax.set_xlabel("Epoch",size=16)
#_=ax.set_yticks(y)
_=ax.set_yticklabels([f"{_:.1f}" for _ in ax.get_yticks()], size=14)
_=ax.set_xticks(x)
_=ax.set_xticklabels(x, size=16)

_=ax.legend(prop={'size':14})

fig.tight_layout()

fig.savefig('plots/sgd_hparams.png')
```
